```python
import numpy as np 
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

%matplotlib inline

def count(series):
    print(Counter(list(series)))
    return Counter(list(series))
def trim(series):
    series.replace('^\s+', '', regex=True, inplace=True) #front
    series.replace('\s+$', '', regex=True, inplace=True) #end
def get_valid_numeric(df, series):
    print(df[pd.to_numeric(df[series], errors='coerce').isnull()][series])
    return df[~pd.to_numeric(df[series], errors='coerce').isnull()]
def create_month_day(df):
    df['Month'] = df['Date'].map(lambda x: x.month)
    df['Day'] = df['Date'].map(lambda x: x.day)
```

```python
df = pd.read_excel('./Pangolin seizures_EIA_030919.xlsx')
print(len(df))
ESTNUM = 'Estimated number of pangolins from whole and/or scales together (for scales, rough estimate of 1 kg scales per 5 kg animal) REFERENCE: COLUMN AB OR COLUMN AE. (If scales co-occur with whole pangolins, judge on case-by-case basis but do not duplicate numbers)'
ESTWEIGHT = 'Estimated weight (kg) - If number of whole pangolins reported but weight not given, estimate 5 kg / pangolin \n(rough estimate; median weight for Asian species)'
SCALESWEIGHT = 'Scales weight (kg). Indicate "Y" if seized, but no weight given'
VIA = 'Via (coming from, before seized)'
SCALESWEIGHT = 'Scales weight (kg). Indicate "Y" if seized, but no weight given'
```

```python
df.columns
```

```python
df[ESTNUM] = df[ESTNUM].replace(r'^\s*$', np.nan, regex=True)
trim(df['Country'])
trim(df['Origin'])
df['Origin'].replace('', 'Unspecified', inplace=True)
df['Origin'].fillna('Unspecified', inplace=True)

trim(df[VIA])
```

```python
df.groupby(['Year', 'Country'])[ESTNUM].sum().unstack().T #.to_csv('./tab-year-country-estnum.csv')
```

```python
df.groupby(['Year'])[ESTNUM].sum()
```

```python
df.groupby(['Country'])[ESTNUM].sum().to_dict()
```

```python
df.groupby(['Year', 'Origin'])[ESTNUM].sum().unstack().T#.to_csv('./tab-year-origin-estnum.csv')
```

## Peak during the year (heatmap)

```python
#df[pd.to_datetime(df['Date'], errors='coerce').isnull()]
```

```python
df_valid_date = df[~pd.to_datetime(df['Date'], errors='coerce').isnull()]
```

```python
df_valid_date['Date'] = pd.to_datetime(df_valid_date['Date'])
```

```python
df_valid_date['Month'] = df_valid_date['Date'].map(lambda x: x.month)
df_valid_date['Day'] = df_valid_date['Date'].map(lambda x: x.day)
```

```python
df_valid_date.groupby(['Month', 'Day'])[ESTNUM].sum().unstack()#.to_csv('./tab-month-day-estnum.csv')
```

## Peak period of Live/Dead/Scales

```python
df_live_date = df_valid_date[pd.notnull(df_valid_date['Live number'])][['Date', 'Live number']]
df_live_date = get_valid_numeric(df_live_date, 'Live number')

df_dead_date = df_valid_date[pd.notnull(df_valid_date['Carcass number'])][['Date', 'Carcass number']]
df_dead_date = get_valid_numeric(df_dead_date, 'Carcass number')

df_scales_date = df_valid_date[pd.notnull(df_valid_date[SCALESWEIGHT])][['Date', SCALESWEIGHT]]
df_scales_date = get_valid_numeric(df_scales_date, SCALESWEIGHT)
```

```python
create_month_day(df_live_date)
create_month_day(df_dead_date)
create_month_day(df_scales_date)
```

```python
df_scales_date.groupby(['Month', 'Day'])[SCALESWEIGHT].sum().unstack().to_csv('./tab-month-day-scalesweight.csv')
```

```python
df_live_date.groupby(['Month', 'Day'])['Live number'].sum().unstack().to_csv('./tab-month-day-livenum.csv')
```

```python
df_dead_date.groupby(['Month', 'Day'])['Carcass number'].sum().unstack().to_csv('./tab-month-day-carcassnum.csv')
```
