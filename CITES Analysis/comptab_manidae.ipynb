## Comparative tabulation of Manidae 
### (aggregated and grouped from web download)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./downloaded/comptab_2019-05-31 12_46_comma_separated.csv')
df.head()
```

#### quantiy match between importer's and exporter's reports

```python
df_quantity_comparable = df[df['Importer reported quantity'].notnull() & df['Exporter reported quantity'].notnull()]
df_quantity_comparable_match = df_quantity_comparable[df_quantity_comparable['Importer reported quantity'] == df_quantity_comparable['Exporter reported quantity']]
df_quantity_comparable_match.head()
```

#### quantity conflict between importer's and exporter's reports

```python
df_quantity_comparable_conflict = df_quantity_comparable[df_quantity_comparable['Importer reported quantity'] != df_quantity_comparable['Exporter reported quantity']]
df_quantity_comparable_conflict
```
