## Convert to Echarts data

```python
from_to_coords_df = pd.read_csv('./china-from-to-coords-fixed.csv')
from_to_coords_df.head()
```

```python
from_to_coords_data = list()
for index, row in from_to_coords_df.iterrows():
    obj = dict()
    obj['fromName'] = row['From']
    obj['toName'] = row['To']
    obj['coords'] = [[row['from_lng'], row['from_lat']],[row['to_lng'], row['to_lat']]]
    from_to_coords_data.append(obj)
```
