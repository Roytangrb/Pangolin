```python
import pandas as pd
import geocoder
import time
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
df = pd.read_csv('fulltext - fulltext.csv')
```

```python
df_court = df[['date', 'court']]
print('total: ', len(df_court))
df_court.head()
```

```python
success_count = 0
for index, address in df_court['court'].iteritems():
    #If there is an error in the connection to the server, 
    #the exception raised by the requests library will be propagated up to the caller. 
    try:
        g = geocoder.baidu(address, key='CpEiE0gRf2ulr1YbkjpbPxhgolTuhX6l')
        #When no result could be found for the given search terms,
        #the ok attribute on the returned object will be False
        if g.ok:
            print(address, g.json['lat'], g.json['lng'])
            df_court.loc[index, 'x'] = g.json['lat']
            df_court.loc[index, 'y'] = g.json['lng']
            df_court.loc[index, 'confidence'] = g.json['confidence']
            success_count = success_count + 1
        else :
            print(address, 'na')
    except:
        print('Exception')
```

```python
df_court.columns = ['year', 'court', 'lat', 'lng', 'confidence']
df_court.head()
```
