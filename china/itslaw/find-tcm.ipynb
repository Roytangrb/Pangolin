## Fulltext analysis(406) - find TCM related cases

```python
import pandas as pd
```

```python
df = pd.read_csv('fulltext - fulltext.csv')
```

```python
df_res = df[df['text'].str.contains('中药|药材|药用|吃药|制药|药厂|药品|药物|服药')].reset_index().drop('index', axis=1)
```

```python
case_count = 0
for text in df_res['text']:
    ps = [p for p in text.split('\n') if ('中药' in p or '药材' in p or '药用' in p or'吃药' in p or'制药' in p or '药厂' in p or '药品'in p or '药物' in p or '服药'in p)]
    if len(ps) > 0:
        case_count = case_count +1
        print(ps)
        print('\n')
```

## Geocode

```python
import geocoder
import time
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
df_court = df_res[['date', 'court']]
df_court.head()
```

```python
success_count = 0
for index, address in df_court['court'].iteritems():
    #If there is an error in the connection to the server, 
    #the exception raised by the requests library will be propagated up to the caller. 
    try:
        g = geocoder.baidu(address, key='CpEiE0gRf2ulr1YbkjpbPxhgolTuhX6l')
        #When no result could be found for the given search terms,
        #the ok attribute on the returned object will be False
        if g.ok:
            print(address, g.json['lat'], g.json['lng'])
            df_court.loc[index, 'x'] = g.json['lat']
            df_court.loc[index, 'y'] = g.json['lng']
            df_court.loc[index, 'confidence'] = g.json['confidence']
            success_count = success_count + 1
        else :
            print(address, 'na')
    except:
        print('Exception')
```

```python
df_court.columns = ['year', 'court', 'lat', 'lng', 'confidence']
df_court
```
