## itslaw fulltext find cases where resturants involves

```python
import pandas as pd
```

```python
df = pd.read_csv('fulltext - fulltext.csv')
```

```python
df_res = df[df['text'].str.contains('餐厅|食用|饭店')]
```

```python
case_count = 0
for text in df_res['text']:
    ps = [p for p in text.split('\n') if ('餐厅' in p or '食用' in p or '饭店' in p)]
    if len(ps) > 0:
        case_count = case_count +1
        print(ps)
        print('\n')
```
