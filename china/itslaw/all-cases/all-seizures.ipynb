### Sum weight and items

```python
import numpy as np
import pandas as pd
from collections import Counter
def count(series):
    s = pd.DataFrame.from_dict(Counter(list(series)), orient='index')
    s.columns = ['count']
    return s
def print_tsv(series):
    for index, value in series.items():
        print('{}\t{}'.format(index, value))
def print_csv(series):
    for index, value in series.items():
        print('{},{}'.format(index, value))
def get_valid_numeric(df, series):
    print(df[pd.to_numeric(df[series], errors='coerce').isnull()][series])
    return df[~pd.to_numeric(df[series], errors='coerce').isnull()]
```

```python
df = pd.read_csv('./joined-379-seizures.csv', dtype={'seizure-year': object, 'seizure-month': object, 'seizure-day': object})
print(df.columns)
print(len(df))
```

```python
print_tsv(count(df['seizure-year'])['count'])
```

### seizure amount in number by year

```python
df_num = pd.read_csv('./joined-seizure-flat-item-value.csv', dtype={'seizure-year': object, 'seizure-month': object, 'seizure-day': object})
df_num = df_num[(pd.notnull(df_num['final value'])) & (pd.notnull(df_num['seizure-year']))]
df_num = get_valid_numeric(df_num, 'final value')
print(len(df_num))
```

```python
print_csv(df_num.groupby('seizure-year').sum()['final value'])
```

### peak period of seizure, by amount
